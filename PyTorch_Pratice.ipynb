Comparing Difference in CPU and GPU to run the same size tensor

```python
import time

# Define the size of the matrices
size = 1000  # Lagre size for performance comparison

# creating random matrices on CPU
martix_cpu1 = torch.rand((size,size))
martix_cpu2 = torch.rand((size,size))

# measures on CPU
start_time = time.time()
result_cpu = torch.matmul(martix_cpu1,martix_cpu2) # matrix multiplication on CPU
end_time = time.time()
cpu_time = end_time - start_time

print(f'Time on CPU: { cpu_time:.4f} seconds')

# move matrices to GPU
martix_gpu1 = martix_cpu1.to('cuda')
martix_gpu2 = martix_cpu2.to('cuda')


# measures on GPU
start_time = time.time()
result_gpu = torch.matmul(martix_gpu1,martix_gpu2) # matix multiplication on GPU
torch.cuda.synchronize() # ensure all GPU are complete
end_time = time.time()
gpu_time = end_time - start_time

print(f'Time on GPU: { gpu_time:.4f} seconds')

# compare results
print('\nSpeedup (CPU time / GPU time):',cpu_time / gpu_time)
```
