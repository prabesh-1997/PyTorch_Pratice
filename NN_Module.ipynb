```python
import torch
import torch.nn as nn

class Model(nn.Module):
    def __init__(self,num_features):

        super().__init__()
        self.network=nn.Sequential(        # Sequintial Container
            nn.Linear(num_features,3),
            nn.ReLU(),
            nn.Linear(3,1),
            nn.Sigmoid()
        )


    def forward(self,features):

          out = self.network(features)
          return out
```

```python
# create dataet
features = torch.rand(10,5)

# create model
model =  Model(features.shape[1])

#call model for forward pass
model(features)
```

```python
# show the weights and bias of the first linear layer
print("Weights of linear1:")
print(model.network[0].weight)
print("Bias of linear1:")
print(model.network[0].bias)

# show the weights and bias of the second linear layer
print("\nWeights of linear2:")
print(model.network[2].weight)
print("Bias of linear2:")
print(model.network[2].bias)
```
